Speech Processing is a multi & inter-disciplinary domain, between physiology, having to know about speech production, hearing and the perception of the human, but also linguistics, having to know about natural language, phonetics & grammar to understand the speech, but also engineering & computer science, having to know about signal processing, machine learning, as well as information coding & transmission.

### Speech Coding
 
Used to transmit as much speech as possible as cheaply as possible (ie. reducing the bitrate). Speech is redundant, meaning we can predict from one sample the next sample with relatively high accuracy, samples are predictable, since the signal has redundancy built into it. And this predictable part is exactly what allows us to reduce the bit rate. Linear prediction is usable in this task.

If the transmission of the speech reaches close to 400ms, it has problems in communciation and it is not usable by humans, so about 200-250ms is the limit to set for effective communication.

### Speech Production

Speech or Sounds are produced from the excitation of folds (vibrating & modulating the air), forming the different sounds.





+ Male pitch is lower than the female pitch, lower than children pitch (related to the fundamental frequency). All depending on differences in vocal fold size, making it heavier, thus vibrating slower, thus reducing the fundamental frequency, making the pitch lower in general.

+ Vocal cords under tension tend to vibrate faster

+ For certain sound productions there is no periodic vibration of vocal cords

+ Length of vocal tract differs from person to person. Male tract is longer than the female tract, longer than the children tract.

+ The shape of the vocal tract changes for different sounds.

### Human Hearing - Perception

+ The Ear transforms the sound's acoustics form of energy into nerve impulses (analog waves -> electrical signals).

+ Having two ears gives us the ability to detect the features of direction when receiving sounds/speech.

+ Loudest sound that can be heard without feeling pain is about 120 dB.

+ The Ear is not equally sensitive to all frequencies (highest sensitivity is for frequencies between 1kHz - 5kHz). This is the reason why we convert the frequencies into the Mel scale (personal info).

+ Having a tone with a certain frequency, it can be masked by a noise. This noise can have different bandwidths, and that can even be a broadband noise, spread across all frequencies. If the bandwidth is defined, there will be a region where weaker signals than the noise will be completely masked.



### Linguistics

***Phoneme*** is the atomic unit of sound, it is the smallest sound unit that distinguishes meaning of words. It is language-specific, and generally represents what our brain perceives and uses to understand meaning.

***Phone*** is several recognizably different sounds of phonemes. It is language-independant, and it generally represents what we physically say

- Each language has a different number of phonemes or phones.

- Phonemes can be grouped in many different ways:

    + Vowels (no constriction in vocal tract) & Consonants.
    + Depending on the manner of articulation (plosive, fricative, approximant).
    + Depending on the place of articulation (labial, dental, alevolar, labio-dental).
    + Palatal-Alevolar, lateral.
    + Whether voiced (periodic vibration of vocal cords, has nonzero fundamental frequency), unvoiced (no fundamental frequency) or aspirant (mix of both) sounds.

- Phonemes is a complete science in of itself, combined with physiology.

## Answer:

1.  - Speech Recognition has the goal of detecting the words being spoken by a human speaker, or rather emulate the human hearing perception of the word, so accumulating the wanted information conveyed by the speaker, including the actual words and their meanings, but also the emotions, intonations and so on. 
    - In Speech Recognition, the input is an acoustics signal of a human speaking, converted into a digital signal, while the output is the words spoken in text.
    - For speech recognition, the challenges are (but not exclusive to) the difference in speech between reading tasks (relatively monotone) and conversation tasks (where we have a huge range of variability), reading tasks are pretty much straight to the point, there are no changes in the tone, nor the speed of reading, while for conversation tasks, the people speaking can speak slowly or fast, they can have accents, their grammar might not be perfectly correct, they might use vocabulary that is not in the "words set". We also have systems that work well with one speaker, but when exposed to another speaker, they do not work as well, so they are speaker-dependant, while if we want the system to be speaker-independant, we would have to detect different accents, different speeds of speech, different pronounciations of the same word, and so on. Multilingual speech recognition is also complicated to achieve, since many languages are under-resourced, and many change over time. Speech recognition also has to be noise-robust, reverberation (echo)-robust, but also robust to both, which is easy to achieve for a human, but not as easy for a machine. It also has to include recognition of atypical speech, whether accented speech, children speech or impaired speech (whether casued by a physical or a mental condition) (a simple cold & autism as examples).
    - While Speech Synthesis is producing speech that is both intelligible & understandable to humans, or rather, emulate the human speaking ability, which involves pronouncing the words right, but also with the right intonation (eg. question) and even with emotion and states that are beyond words. 
    - For Speech Synthesis, the input is text, and the output are acoustic signal of a human-like voice speaking.
    - For Speech synthesis, the challenges are pronouncing the words right, affective/expressive speech synthesis, where a person would be able to recognize whether the sentence being said is a question or a statement, producing multilingual speech by the same "voice", this voice sometimes becomes unseparable (the case of Stephen Hawkins) where the person cannot change the voice since it became a part of their identity. It also has to perform well in "objective evaluations", meaning that people understand what is being said, and they judge it as "natural" as possible.
    - Coarticulation is in simple terms the way in which the pronunciation of a speech sound is influences by the sounds that come before and after it. This is what causes the "blending" of sounds, making our speech fluid and natural rather than a series of disjointed phonemes. It is the way our articulators (tongue, lips, jaws, etc..) anticipating the next sound and carry over characteristics from the previous one. Coarticulation is desirable in Speech Synthesis, since it is essential for producing natural and human-like speech. A high-quality TTS system must accurately model how phonemes blend together. For Speech Recognition, coarticulation is a significant challenge, since it introduces too much variability into the speech signal.

2. - There are two primary methods for determining the phone sequences for words in a language:
        + Pronunciation Dictionaries (Lexicons): being the most common method, it consists in using a specially currated dictionary that maps words to their phonetics representations. These dictionaries are often created by linguists and phoneticians who manually transcribe words (eg. CMU Pronouncing Dictionary).
        + Grapheme-to-Phoneme (G2P) Conversion: for words that are not in a dictionary (out-of-vocabulary words), systems relu on a G2P converter. This is a model, often using ML, that learns the relationship between the spelling of a word (its graphemes) and its pronunciation (its phonemes). While a language like English has many irregular spellings (eg. "through", "tough", "though"), a G2P system can learn these complex rules from data and predict the most likely pronunciation for a new word.

    - No, a living language can ***never*** have a truly "closed set" vocabulary. The set of distinct words in any actively used language is constantly changing. There are many reasons for this, including: 
        + **Neologisms** (New words): humans are constantly innovating and creating new concepts, requiring new words. Words like "selfie" or "google" (as a verb) did not exist a few decades ago but are now part of our common vocabulary.
        + **Borrowing from Other Languages**: Languages are not isolated. Through cultural exchange, trade, and migration, words are constantly being borrowed from other languages. In English for example, there are a lot of words borrowed from other languages (eg., "deja vu" from French, "kindergarten" from German).
        + **Slang and Idiomatic Evolution**: Language evolves organically. Slang terms emerge and can eventually become part of the standard language. The meaning of existing words can also shift or expand over time.

        Because language is a dynamic and evolving tool for human communication, its vocabulary will always remain an **open set**, continuously adapting to the changing needs and experiences of its speakers.
